```python
import numpy as np
import sys
class Node:
    def __init__(self,value,point):
        self.value = value
        self.point = point
        self.parent = None
        self.H = 0
        self.G = 0
    def move_cost(self,other):
        return 0 if self.value == '.' else 1
        
def children(point,grid):
    listt = set()
    x,y = point.point
    rows = len(grid)
    colomns = len(grid[0])
    if x == 0 and y==0:
        links = set([grid[d[0]][d[1]] for d in [(x+1,y), (x,y + 1)]])
        #printt([link for link in links if links[link] != '%'])
        for link in links:
            if link.value != '%':
                listt.add(link)
        return listt
    elif x==rows-1 and y==0:
        links = set([grid[d[0]][d[1]] for d in [(x,y + 1), (x-1, y)]])
        #print([link for link in links if links[link] != '%'])
        for link in links:
            if link.value != '%':
                listt.add(link)
        return listt
    elif x == 0 and y==colomns-1:
        links = set([grid[d[0]][d[1]] for d in [(x+1,y), (x,y - 1)]])
        #print([link for link in links if links[link] != '%'])
        for link in links:
            if link.value != '%':
                listt.add(link)
        return listt
    elif x==rows-1 and y ==colomns-1:
        links = set([grid[d[0]][d[1]] for d in [(x,y - 1), (x-1, y)]])
        #print([link for link in links if links[link] != '%'])
        for link in links:
            if link.value != '%':
                listt.add(link)
        return listt
    #first row
    elif x==0 :
        links = set([grid[d[0]][d[1]] for d in [(x+1,y), (x,y+1), (x,y - 1)]])
        #print([link for link in links if links[link] != '%'])
        for link in links:
            if link.value != '%':
                listt.add(link)
        return listt        
    #last row
    elif x==rows-1:
        links = set([grid[d[0]][d[1]] for d in [(x,y+1), (x,y - 1), (x-1, y)]])
        #print([link for link in links if links[link] != '%'])
        for link in links:
            if link.value != '%':
                listt.add(link)
        return listt 
    #first colomn
    elif y == 0:    
        links = set([grid[d[0]][d[1]] for d in [(x-1,y), (x,y+1), (x+1, y),]])
        #print([link for link in links if links[link] != '%'])
        for link in links:
            if link.value != '%':
                listt.add(link)
        return listt
    #last colomn
    elif y == colomns-1: 
        links = set([grid[d[0]][d[1]] for d in [ (x-1,y), (x,y - 1), (x+1, y)]])
        #print([link for link in links if links[link] != '%'])
        for link in links:
            if link.value != '%':
                listt.add(link)
        return listt
    
    else:
        links = set([grid[d[0]][d[1]] for d in [(x+1,y), (x,y+1), (x,y - 1), (x-1, y) ]])
        #print([link for link in links if links[link] != '%'])
        for link in links:
            if link.value != '%':
                listt.add(link)
        return listt

def manhattan(point,point2):
    return abs(point.point[0] - point2.point[0]) + abs(point.point[1]-point2.point[0])
def aStar(start, goal, grid):
    #The open and closed sets
    openset = set()
    closedset = set()
    #Current point is the starting point
    current = start
    #Add the starting point to the open set
    openset.add(current)
    #While the open set is not empty
    while openset:
        #Find the item in the open set with the lowest G + H score
        current = min(openset, key=lambda o:o.G + o.H)
        #If it is the item we want, retrace the path and return it
        if current == goal:
            path = []
            while current.parent:
                path.append(current)
                current = current.parent
            path.append(current)
            return path
        #Remove the item from the open set
        openset.remove(current)
        #Add it to the closed set
        closedset.add(current)
        #Loop through the node's children/siblings
        #print((children(current,grid)))
        chlist = children(current,grid)
        for node in chlist:
            #If it is already in the closed set, skip it
            if node in closedset:
                continue
            #Otherwise if it is already in the open set
            if node in openset:
                #Check if we beat the G score 
                new_g = current.G + current.move_cost(node)
                if node.G > new_g:
                    #If so, update the node to have a new parent
                    node.G = new_g
                    node.parent = current
            else:
                #If it isn't in the open set, calculate the G and H score for the node
                node.G = current.G + current.move_cost(node)
                node.H = manhattan(node, goal)
                #Set the parent to our current item
                node.parent = current
                #Add it to the set
                openset.add(node)
#     #Throw an exception if there is no path
#     raise ValueError('No Path Found')
def next_move(pacman_x, pacman_y, food_x, food_y, grid):
    #Convert all the points to instances of Node
    for x in range(len(grid)):
        for y in range(len(grid[x])):
            grid[x][y] = Node(grid[x][y],(x,y))
    #Get the path
    path = aStar(grid[pacman_x][pacman_y],grid[food_x][food_y],grid)
    #Output the path
    #print len(path) - 1
    for node in path[::-1]:
        x, y = node.point
        print (x, y)
pacman_x, pacman_y = [ int(i) for i in input().strip().split() ]
food_x, food_y = [ int(i) for i in input().strip().split() ]
x,y = input().strip().split()
x=int(x)
y=int(y)
 

# for i in range(0, x):
#     grid.append(list(input().strip()))
matrix = []
for j in range(0,int(x)):
    matrix.append(list(map(str, input().rstrip().split()))) 

grid = list()
for j in range(0,int(x)):
    l=[]
    for i in range (0,y):
        l.append(str(matrix[j])[i+2])
    grid.append(l)
    
next_move(pacman_x, pacman_y,food_x, food_y, grid)
```
