```python
def gradient_update(f, x0, y0, alpha):
    
    f0 = f(x0, y0) 
    #to define x+e for deriviation formula
    x = x0 + 0.00000001
    f1 = f(x, y0) 
    gx = (f1-f0)/0.00000001
    #update x
    new_x = x0 - alpha * gx
    
    #to define y+e for deriviation formula
    y = y0 + 0.00000001
    f1 = f(x0, y) 
    gy = (f1-f0)/0.00000001
    #update y
    new_y = y0 - alpha * gy    
    
    return new_x, new_y
    
    
def gd(f, initial, alpha, thereshold, steps):
    
    x_points = []
    y_points = []
    x = initial[0]
    y = initial[1]
    x_points.append(x)
    y_points.append(y)
    step = 1
    while step <=steps:
        new_x, new_y = gradient_update(f, x, y, alpha)
        # add points to the list 
        x_points.append(new_x)
        y_points.append(new_y)
        #check treshholds
        if np.abs(new_x) > thereshold or np.abs(new_y) > thereshold:
            return x_points, y_points
        x = new_x
        y = new_y
        step +=1
        
    return x_points, y_points
```

            \end{cases}$$
            که n تعداد کل داده‌ها است.

            ب)
            تابع زیر را کامل کنید.
            این تابع
            مقدار لحظه‌ای بردار w
            مجموعه‌ی داده‌ی U
            و همچنین Z
            را ورودی می‌گیرید.
            Z 
            شامل آرایه‌ای از برچسب‌های داده‌ها است.
            U
            نیز به صورت آرایه‌ای به فرم
            $U[i] = (x_i, y_i, 1)$
            ورودی داده می‌شود.
            همچنین G نیز در ورودی تابع داده می‌شود.

```python
def compute_loss(w, G, U, Z):
    
    w_norm = (np.linalg.norm(w))**2
    dot =np.dot(U, w)
    summ = 1 - (Z.T * dot)
    summ[summ<0] = 0   #MAX
    N = U.shape[0]   #FINDE THE NUMBER OF DAATA
    loss = 1/2 * w_norm + G * np.sum(summ)/N

    return loss
```

```python
def compute_gradient(w, G, u, z):

    w_norm = (np.linalg.norm(w))**2
    dot =np.dot(u,w)
    summ = 1 - z * dot
    if summ <=0 :
        gradient = w
    else:
        gradient = w - G *z *u
    
    return gradient
```
